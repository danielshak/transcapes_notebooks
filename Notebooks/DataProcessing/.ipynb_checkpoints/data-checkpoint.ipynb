```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

### Table containing all of the rna-seq data for both cell line and all ages, includes ~40k genes. Need to remove zero rows

```python
rna_seq = pd.read_csv('../../Data/RawData/PPG_RNAseq_covariance_input_0419.tsv',sep='\t',index_col=0)
rna_seq.head()
```

### Proteomics Data

```python
proteomics_AM = pd.read_csv('../../Data/RawData/PPG_AM_proteomics_covariance_input_0419.tsv',sep='\t',index_col=0)
proteomics_AM.columns = ['AvgChrs','AM_18M_F0_P','AM_04M_F0_P','AM_18M_F10_P','AM_04M_F10_P']
proteomics_AM.drop(columns='AvgChrs',inplace=True)
print(proteomics_AM.head())
print(proteomics_AM.shape)

proteomics_AT = pd.read_csv('../../Data/RawData/PPG_AT2_proteomics_covariance_input_0419.tsv',sep='\t',index_col=0)
proteomics_AT.columns = ['AvgChrs','AT2_18M_F0_P','AT2_04M_F0_P','AT2_18M_F10_P','AT2_04M_F10_P']
proteomics_AT.drop(columns='AvgChrs',inplace=True)
print(proteomics_AT.head())
print(proteomics_AT.shape)
```

### Reference chromosome lengths

```python
chrom_info = pd.read_csv('../../Data/RawData/ChromosomeLength_MmG38p6.tsv',sep='\t',index_col=0)
chrom_info
```

### Reference start/end locations as well as the chromosome number

```python
sal_EnsembleIds_StartEnd = pd.read_csv('../../Data/RawData/mart_export_mm_gr38_82718.txt',sep='\t',index_col=0)
sal_EnsembleIds_StartEnd.head()
```

### Combine rna-seq with gene start/end and lengths

```python
assert all([gene in sal_EnsembleIds_StartEnd.index for gene in rna_seq.index]) #ensures that all of the gene indicie match
rna_seq = rna_seq.join(sal_EnsembleIds_StartEnd)

#Add gene lenth
gene_length = pd.DataFrame(rna_seq['Gene end (bp)'] - rna_seq['Gene start (bp)'],columns=['Gene Length'])
rna_seq = rna_seq.join(gene_length)
rna_seq.head()
```

### Combine rna-seq with proteomics

```python
rna_seq = rna_seq.join(proteomics_AM,on='Gene.names')
rna_seq = rna_seq.join(proteomics_AT,on='Gene.names')
rna_seq.fillna(0,inplace=True)
rna_seq.head()
```

### Recalculate avg chrm position

```python
chrom_lengths = dict(chrom_info['length'])
chrom_nums = {val: num for num,val in enumerate(chrom_info.index)}

avg_position = ((rna_seq['Gene end (bp)'] + rna_seq['Gene start (bp)'])/2/ rna_seq['Chromosome/scaffold name'].map(chrom_lengths))
NormPosition = pd.DataFrame(avg_position,columns=['NormPosition'])
chrom_pos = pd.DataFrame(rna_seq['Chromosome/scaffold name'].map(chrom_nums) + avg_position,columns=['AvgChrs2'])

rna_seq = rna_seq.join(NormPosition)
rna_seq = rna_seq.join(chrom_pos)

if max(rna_seq['AvgChrs'] - rna_seq['AvgChrs2']) < 10e-6:
    rna_seq.drop(columns='AvgChrs2',inplace=True)
```

### Remove genes with no values for both rna-seq or proteomics

```python
cell_lines_rna = ['AT2_04M_F0','AT2_18M_F0','AT2_04M_F10','AT2_18M_F10','AM_04M_F0','AM_18M_F0','AM_04M_F10','AM_18M_F10']
cell_lines_prot = ['AT2_04M_F0_P','AT2_18M_F0_P','AT2_04M_F10_P','AT2_18M_F10_P','AM_04M_F0_P','AM_18M_F0_P','AM_04M_F10_P','AM_18M_F10_P']
cell_lines = cell_lines_rna+cell_lines_prot

#Removes rows where all rna-seq values and all protein values are 0
rna_seq = rna_seq[rna_seq[cell_lines].sum(axis=1)!=0]
```

### Missing protein values with known rna-seq values

```python
missing_prot = rna_seq[rna_seq[cell_lines_prot].sum(axis=1)==0]
```

### Missing rna-seq with known protein values

```python
missing_rna = rna_seq[rna_seq[cell_lines_rna].sum(axis=1)==0]
```

### Remove genes with missing rna-seq

```python
GP_data = rna_seq[rna_seq[cell_lines_rna].sum(axis=1)!=0]
```

### Saving tables

```python
missing_prot.to_csv(path_or_buf='../../Data/ProcessedData/missing_prot.tsv',sep='\t')
missing_rna.to_csv(path_or_buf='../../Data/ProcessedData/missing_rna.tsv',sep='\t')
GP_data.to_csv(path_or_buf='../../Data/ProcessedData/GP_data.tsv',sep='\t')
```
