### Notebook to load and combine unirep protein embedding chunks
* Chunks are dictionaries containing the original gene name, protein sequence, unirep output (avg hidden, final hidden, and final cell representation)
* Saves a single dictionary containing all of the chunked information

```python
import pickle
import pandas as pd
```

```python
chunks_directory = '../../Data/missing_proteins/protein_embeddings'
ppg_missing_chunk0_embeddings64
```

```python
df_chunks = []
for i in range(30):
    with open(f'{chunks_directory}/ppg_missing_chunk{i}_embeddings64.pkl','rb') as file:
        chunk = pickle.load(file)
        
    df_chunks.append(pd.DataFrame.from_dict(chunk,orient='index',columns=['protein_sequence','unirep_embedding']))
    
protein_embeddings = pd.concat(df_chunks)
protein_embeddings[['avg_embedding','final_hidden','final_cell']] = pd.DataFrame(protein_embeddings.unirep_embedding.values.tolist(),
                                                                                 index= protein_embeddings.index)
protein_embeddings.drop(columns=['unirep_embedding'],inplace=True)

#Below might have to change depending on namming
protein_embeddings.index.name = 'UniParc ID'
protein_embeddings.index = protein_embeddings.index.map(lambda x: x.split()[0])
protein_embeddings
```

```python
#slow
protein_embeddings.to_csv('../../Data/missing_proteins/uniparc_embeddings.tsv',sep='\t')
```

```python
with open('../../Data/missing_proteins/uniparc_embeddings.pkl', 'wb') as file:
    pickle.dump(protein_embeddings, file, protocol=pickle.HIGHEST_PROTOCOL)
```
