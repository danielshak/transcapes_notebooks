````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Bio.SeqUtils
from collections import Counter
import sys
sys.path.insert(0,'/gpfs/home/dshak/Reference_Data')
from rna_codons import CODONS
from tqdm import tqdm
````

````python
data = pd.read_csv('../../29atlas/EV_tables_quantification/Table_EV3/Table_EV3.tsv',sep='\t')
````

## Contains the mRNA and protein levels (ibaq) in log10

````python
adrenal = data[['GeneName','EnsemblGeneID','EnsemblTranscriptID','EnsemblProteinID','Adrenal_mRNA', 'Adrenal_protein','Adrenal_PTR']]
adrenal = adrenal[adrenal['Adrenal_mRNA']!='   NA']
adrenal.set_index('GeneName',inplace=True)
adrenal.head()
````

## Sequences

````python
sequences = pd.read_csv('../../29atlas/EV_tables_quantification/Table_EV4/Table_EV4.tsv',sep='\t')
sequences.set_index('GeneName',inplace=True)
sequences.head()
````

## Merge Seqences, mRNA/protein numbers, and add columns for codons and proteins

````python
aminoacid_keys = list(set(CODONS.values()) - set('_'))
codon_keys = list(set(CODONS.keys()))
````

````python
adrenal = adrenal.join(sequences[['UTR5_Sequence', 'CDS_Sequence', 'UTR3_Sequence','ProteinSequence']])
cols = list(set(CODONS.values()) - set('_')) + list(set(CODONS.keys()))
codons_proteins = pd.DataFrame(columns = cols)
adrenal = adrenal.join(codons_proteins).fillna(0)

adrenal.head()
````

### Reference for gene length and chrm position

````python
ref = pd.read_csv('../../29atlas/ref_genome/grch38v83_gene_st_end_chrm.txt',sep='\t')
#ref = ref.rename(columns = {'Ensembl Gene ID':'EnsemblGeneID'})
ref.set_index('Ensembl Transcript ID',inplace=True)
ref.head()
````

### Gene start/end, lengths, transcript counts

````python
#Do the match on the transcripts in the reference since there are multple transcripts in the reference for the same gene
#match the transcript in data to ref
adrenal = adrenal.join(ref[['Gene Start (bp)', 'Gene End (bp)', 'Chromosome Name']],on='EnsemblTranscriptID')
adrenal['GeneLength'] = adrenal['Gene End (bp)']-adrenal['Gene Start (bp)']
adrenal['CDS_length'] = adrenal['CDS_Sequence'].str.len()
adrenal['UTR5_length'] = adrenal['UTR5_Sequence'].str.len()
adrenal['UTR3_length'] = adrenal['UTR3_Sequence'].str.len()

trans_counts = pd.DataFrame.from_dict(dict(Counter(list(ref['Ensembl Gene ID']))),orient='index',columns=['transcript_counts'])
adrenal = adrenal.join(trans_counts,on='EnsemblGeneID')
adrenal.head()
````

### Calculates the codon frequency and amino acid frequency

````python
n = 3

for gene in tqdm(adrenal.index):
    seq = adrenal.loc[gene,'CDS_Sequence']
    prt = adrenal.loc[gene,'ProteinSequence']
    codon_counts = Counter([seq[i:i+n] for i in range(0, len(seq),n)])
    prt_counts = Counter(list(prt))
    
    num_codons = sum(codon_counts.values())
    num_prts = sum(prt_counts.values())
    
    adrenal.loc[gene,codon_counts.keys()] = [count/num_codons for count in codon_counts.values()]
    adrenal.loc[gene,prt_counts.keys()] = [count/num_prts for count in prt_counts.values()]
````

### GC content of cds,3' and 5'

````python
cds_gc = adrenal.loc[:,'CDS_Sequence'].str.count('G|C')/adrenal.loc[:,'CDS_Sequence'].str.len()
utr5_gc = adrenal.loc[:,'UTR5_Sequence'].str.count('G|C')/adrenal.loc[:,'UTR5_Sequence'].str.len()
utr3_gc = adrenal.loc[:,'UTR3_Sequence'].str.count('G|C')/adrenal.loc[:,'UTR3_Sequence'].str.len()
gc_df = pd.concat([cds_gc,utr5_gc,utr3_gc],axis='columns')
gc_df.columns = ['cds_gc','utr5_gc','utr3_gc']
adrenal = adrenal.join(gc_df)
adrenal.head()
````

### Molecular Mass Calculations

````python
mm = adrenal.apply(lambda x: Bio.SeqUtils.molecular_weight(x['ProteinSequence'],seq_type = 'protein'), axis = 1)
mm.name = 'MolecularMass'
adrenal = adrenal.join(mm)
````
