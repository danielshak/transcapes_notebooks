```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

current_dir = os.getcwd()
data_dir = os.path.join(current_dir, '../../Data/')

ratios_path = data_dir+'29Atlas/QuantificationTables/Table_EV3/Table_EV3.tsv'
```

```python
def scatter_genes(series,save=False):
    fig = plt.figure()
    fig.set_size_inches(10.5,10.5)
    ax = fig.add_subplot(111)
    ax.scatter(series[0:29].astype(float),series[29::].astype(float),
               marker='o',alpha=0.7,s=120,edgecolors='black',c='gray',linewidths=1.5)
    ax.set_xlim(1,3)
    ax.set_ylim(5,9)
    ax.set_xlabel('mRNA fpkm log10', fontsize=18)
    ax.set_ylabel('Protein ibaq log10', fontsize=18)
    ax.set_title(f'Gene {series.name} Measurments Across Tissues', fontsize=20)
    if save == True:
        fig.savefig(f'./figures/scatter_genes/{series.name}_scatter.pdf')
```

```python
gene_table = pd.read_csv(ratios_path,sep='\t',index_col=0,usecols=[0]+list(range(4,62)))
```

```python
counts = dict()
transposed = gene_table.T
for name in transposed.columns:
    counts[name] = transposed[name].str.contains('NA').sum()

counted_genes = pd.DataFrame.from_dict(counts,orient='index')
sorted_genes = counted_genes.sort_values(by=0)
sorted_genes.head(20:30)
```
