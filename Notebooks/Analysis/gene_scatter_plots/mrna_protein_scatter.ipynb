```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import numpy as np
import os
import sys

sys.path.append('../../Utils')
from data_class import GPY_data

current_dir = os.getcwd()
data_dir = os.path.join(current_dir, '../../../Data/')

ratios_path = data_dir+'ProcessedData/GP_data.tsv'

#Cell lines
AT2 = ['AT2_04M_F0','AT2_04M_F10','AT2_18M_F0','AT2_18M_F10']
AM = ['AM_04M_F0','AM_04M_F10','AM_18M_F0','AM_18M_F10']
cell_lines = AT2+AM
```

```python
gene_table = pd.read_csv(ratios_path,sep='\t',index_col=0,usecols=[1]+list(range(4,12))+list(range(16,24)))
gene_table
```

```python
for name in cell_lines:
    missing_pairs = (gene_table.loc[:,[name,f'{name}_P']]==0).any(axis=1)
    print(f'{name} is missing {missing_pairs.sum()} pairs contains {15255-missing_pairs.sum()} pairs')
    gene_table.loc[missing_pairs,[name,f'{name}_P']] = 0 
```

```python
gene_pairs = gene_table[(gene_table == 0).sum(axis=1)!=16] #4694 rows
sorted_genes = (gene_pairs != 0).sum(axis=1).sort_values(ascending=False).index
sorted_gene_pairs = gene_pairs.reindex(sorted_genes)
```

```python
def scatter_genes(series,cell_lines,save=False):
    fig = plt.figure()
    fig.set_size_inches(10.5,10.5)
    ax = fig.add_subplot(111)
    
    scatter_symbols = ['o','v','s','X']
    syms_legend_handles = []
    
    for sym,at,am in zip(scatter_symbols,cell_lines[0:4],cell_lines[4:8]):
        if series[at] > 0:
            ax.scatter(series[at],series[f'{at}_P'],marker=sym,alpha=0.7,s=120,
                      edgecolors='black',c='red',linewidths=1.5)
        if series[am] > 0:
            ax.scatter(series[am],series[f'{am}_P'],marker=sym,alpha=0.7,s=120,
                      edgecolors='black',c='blue',linewidths=1.5)
        
        syms_legend_handles.append(mlines.Line2D([], [], color='black', marker=sym,
                          markersize=10, label=''.join(at.split('_')[-2::]),linestyle='None'))
    ax.set_xlim(1,11) #+- ~3 standard deviations
    ax.set_ylim(20,29) #+- ~3 standard deviations
    ax.set_xlabel('mRNA tmm log2', fontsize=18)
    ax.set_ylabel('Protein AUC log2', fontsize=18)
    ax.set_title(f'Gene {series.name} Measurments Across Cell Conditions', fontsize=20)
    
    #Setting up legends
    red = mpatches.Patch(color='red', label='AT2 Cells')
    blue = mpatches.Patch(color='blue', label='AM Cells')
    color_legend = plt.legend(handles=[red,blue],loc='lower left')
    ax.add_artist(color_legend)
    
    ax.legend(handles=syms_legend_handles,loc='lower right')
    
    if save == True:
        fig.savefig(f'./plots/{series.name}_scatter.pdf')
        plt.close(fig)
```

```python
#first n genes in sorted list
n=100
for i in range(n):
    scatter_genes(sorted_gene_pairs.iloc[i,:],cell_lines,save=True)
```
